```python
# Author: Pankaj Chouhan
# Website: www.codeswithpankaj.com
# Tutorial: Codes With Pankaj Chouhan

import numpy as np
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

# Sample dataset: Hours Studied and Exam Scores
data = {
    'Hours_Studied': [2, 3, 5, 6, 8, 1, 4, 7],
    'Exam_Score': [50, 55, 70, 80, 90, 45, 60, 85]
}

# Create a DataFrame (like a spreadsheet in Python)
df = pd.DataFrame(data)

# Calculate Pearson Correlation Coefficient and P-value
correlation, p_value = pearsonr(df['Hours_Studied'], df['Exam_Score'])

# Print the dataset and results
print("Dataset Preview:")
print(df)
print("\nPearson Correlation Coefficient (r):", round(correlation, 3))
print("P-value:", round(p_value, 4))

# Interpret the correlation
print("\nCorrelation Interpretation:")
if correlation > 0.7:
    print("Strong positive linear relationship.")
elif 0.3 < correlation <= 0.7:
    print("Moderate positive linear relationship.")
elif 0 < correlation <= 0.3:
    print("Weak positive linear relationship.")
elif correlation == 0:
    print("No linear relationship.")
elif -0.3 <= correlation < 0:
    print("Weak negative linear relationship.")
elif -0.7 <= correlation < -0.3:
    print("Moderate negative linear relationship.")
else:
    print("Strong negative linear relationship.")

# Interpret the p-value
print("\nP-value Interpretation:")
if p_value < 0.05:
    print("The correlation is statistically significant (not due to chance).")
else:
    print("The correlation is not statistically significant (might be random).")

# Visualize with a scatter plot
plt.scatter(df['Hours_Studied'], df['Exam_Score'], color='blue', label='Data Points')
plt.plot(np.unique(df['Hours_Studied']), 
         np.poly1d(np.polyfit(df['Hours_Studied'], df['Exam_Score'], 1))(np.unique(df['Hours_Studied'])), 
         color='red', label='Trend Line')
plt.title('Hours Studied vs Exam Score')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```
