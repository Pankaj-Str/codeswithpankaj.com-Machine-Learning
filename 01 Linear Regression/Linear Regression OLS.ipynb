```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Data (reshape X for sklearn)
hours = np.array([2, 3, 5, 7, 8]).reshape(-1, 1)  # Study hours
scores = np.array([20, 30, 50, 70, 80])  # Exam scores

# Build and train the model
model = LinearRegression()
model.fit(hours, scores)

# Get slope and intercept
beta_1 = model.coef_[0]
beta_0 = model.intercept_

print(f"Slope (β₁): {beta_1:.2f}")
print(f"Intercept (β₀): {beta_0:.2f}")

# Predictions
predicted_scores = model.predict(hours)

# Plot it!
plt.scatter(hours, scores, color="blue", label="Actual Scores")
plt.plot(hours, predicted_scores, color="red", label="Best-Fit Line")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Linear Regression with Scikit-learn")
plt.legend()
plt.show()

# Predict for 6 hours
new_hours = np.array([[6]])
new_score = model.predict(new_hours)[0]
print(f"Predicted score for 6 hours: {new_score:.2f}")
```

```python
from sklearn.metrics import r2_score, mean_squared_error

r2 = r2_score(scores, predicted_scores)
mse = mean_squared_error(scores, predicted_scores)

print(f"R-squared: {r2:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
```
